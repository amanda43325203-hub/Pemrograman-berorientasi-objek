```python
class Student:
    def __init__(self, name, score):
        self.__name = name
        self.__score = 0
        self.__grade = ""
        self.score = score   # langsung lewat setter biar kecek

    # getter name
    @property
    def name(self):
        return self.__name

    # setter name
    @name.setter
    def name(self, value):
        if value == "":
            print("Nama ga boleh kosong")
        else:
            self.__name = value

    # getter score
    @property
    def score(self):
        return self.__score

    # setter score
    @score.setter
    def score(self, value):
        if value < 0 or value > 100:
            print("Nilai harus 0 - 100")
        else:
            self.__score = value
            self.__hitung_grade()

    # fungsi buat nentuin grade
    def __hitung_grade(self):
        if self.__score >= 90:
            self.__grade = "A"
        elif self.__score >= 80:
            self.__grade = "B"
        elif self.__score >= 70:
            self.__grade = "C"
        elif self.__score >= 60:
            self.__grade = "D"
        else:
            self.__grade = "E"

    # tampilkan data
    def show_info(self):
        print("Nama Mahasiswa :", self.__name)
        print("Nilai :", self.__score)
        print("Grade :", self.__grade)

    # destruktor
    def __del__(self):
        print(f"Data mahasiswa {self.__name} sudah dihapus")


# contoh jalanin program
if __name__ == "__main__":
    mhs = Student("Amandaa", 98)
    mhs.show_info()

    print("\nNilai diubah...\n")
    mhs.score = 93
    mhs.show_info()

    del mhs
```
