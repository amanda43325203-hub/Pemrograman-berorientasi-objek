### Praktikum 01: Polimorfisme dengan Inheritance dan Overriding 


```python
class Burung:
    def __init__(self, nama):
        self.nama = nama

    def terbang(self):
        print(f"{self.nama} terbang dengan cara umum.")

    def bersuara(self):
        print(f"{self.nama} mengeluarkan suara burung.")


class Elang(Burung):
    def __init__(self, nama, rentang_sayap):
        super().__init__(nama)
        self.rentang_sayap = rentang_sayap

    def terbang(self):
        print(f"{self.nama} terbang tinggi melayang di angkasa.")

    def bersuara(self):
        print(f"{self.nama} berteriak nyaring!")


class Pipit(Burung):
    def __init__(self, nama, warna_bulu):
        super().__init__(nama)
        self.warna_bulu = warna_bulu

    def terbang(self):
        print(f"{self.nama} terbang cepat di antara pepohonan.")

    def bersuara(self):
        print(f"{self.nama} berkicau merdu: Cit cit!")


def demonstrasi_aksi_burung(daftar_burung):
    print("\nAksi Burung:")
    for burung in daftar_burung:
        print(f"-- Aksi untuk {burung.nama} --")
        burung.terbang()
        burung.bersuara()
        print("-" * 15)


if __name__ == "__main__":
    elang = Elang("Elang Jawa", 1.5)
    pipit = Pipit("Pipit Gereja", "Coklat")
    burung = Burung("Burung Misterius")

    daftar = [elang, pipit, burung]
    demonstrasi_aksi_burung(daftar)
```

### Praktikum 02: Polimorfisme dengan Duck Typing 


```python
class Kucing:
    def bersuara(self):
        print("Kucing: Meow!")

class Anjing:
    def bersuara(self):
        print("Anjing: Guk guk!")

class Bebek:
    def bersuara(self):
        print("Bebek: Kwek kwek!")

class Mobil:
    def jalankan(self):
        print("Mobil: Brummm!")


def buat_suara(obj):
    try:
        obj.bersuara()
    except AttributeError:
        print(f"Objek {type(obj).__name__} tidak bisa bersuara.")


if __name__ == "__main__":
    daftar = [Kucing(), Anjing(), Bebek(), Mobil()]

    for item in daftar:
        buat_suara(item)
```

### Praktikum 03: Simulasi Overloading dalam Metode Kelas dengan *args 


```python
class KalkulatorSederhana:
    def __init__(self, nama="Kalkulator"):
        self.nama = nama
        print(f"{self.nama} siap digunakan.")

    def tambah(self, *args):
        if not args:
            print("Tidak ada angka.")
            return 0

        total = 0
        for angka in args:
            if isinstance(angka, (int, float)):
                total += angka
            else:
                print(f"{angka} bukan angka")

        print(f"Hasil: {total}")
        return total


if __name__ == "__main__":
    calc = KalkulatorSederhana()

    calc.tambah(5, 10)
    calc.tambah(1, 2, 3, 4)
    calc.tambah(100)
    calc.tambah()
```

### Praktikum 04: Simulasi Overloading dalam Metode Kelas dengan *args dan **kwargs 


```python
class DataProcessor:
    def __init__(self, id):
        self.id = id
        print(f"Processor {self.id} siap.")

    def process(self, *args, **kwargs):
        print("\nProses Data")

        if args:
            print("Args:")
            for a in args:
                print(a)
        else:
            print("Tidak ada args")

        if kwargs:
            print("Kwargs:")
            for k, v in kwargs.items():
                print(k, ":", v)
        else:
            print("Tidak ada kwargs")


if __name__ == "__main__":
    dp = DataProcessor("DP-01")

    dp.process()
    dp.process(1, 2, 3)
    dp.process(nama="Amanda", umur=20)
    dp.process(10, 20, status="OK")
```

### Praktikum 05: Kombinasi Polimorfisme (Inheritance & Duck Typing) 


```python
import math

class Bentuk:
    def nama_bentuk(self):
        return "Bentuk"

    def hitung_luas(self):
        raise NotImplementedError


class Lingkaran(Bentuk):
    def __init__(self, r):
        self.r = r

    def nama_bentuk(self):
        return "Lingkaran"

    def hitung_luas(self):
        return math.pi * self.r**2


class Persegi(Bentuk):
    def __init__(self, s):
        self.s = s

    def nama_bentuk(self):
        return "Persegi"

    def hitung_luas(self):
        return self.s * self.s


class TaplakMeja:
    def __init__(self, p, l):
        self.p = p
        self.l = l

    def hitung_luas(self):
        return self.p * self.l


def tampilkan(obj):
    try:
        luas = obj.hitung_luas()
        try:
            nama = obj.nama_bentuk()
        except:
            nama = type(obj).__name__

        print(nama, ":", luas)
    except:
        print("Tidak bisa hitung luas")


if __name__ == "__main__":
    data = [
        Lingkaran(7),
        Persegi(5),
        TaplakMeja(2, 3),
        Bentuk(),
        "string"
    ]

    for item in data:
        tampilkan(item)
```

### Praktikum 06: Kontrol Perilaku Polimorfik dengan isinstance() 


```python
class Burung:
    def __init__(self, nama):
        self.nama = nama

    def terbang(self):
        print(f"{self.nama} terbang.")

    def bersuara(self):
        print(f"{self.nama} bersuara.")


class Elang(Burung):
    def terbang(self):
        print(f"{self.nama} terbang tinggi.")

    def bersuara(self):
        print(f"{self.nama} nyaring!")

    def berburu(self):
        print(f"{self.nama} berburu.")


class Pipit(Burung):
    def terbang(self):
        print(f"{self.nama} terbang cepat.")

    def bersuara(self):
        print(f"{self.nama} cit cit!")

    def membangun_sarang(self):
        print(f"{self.nama} buat sarang.")


def interaksi(b):
    if isinstance(b, Burung):
        b.terbang()
        b.bersuara()

        if isinstance(b, Elang):
            b.berburu()
        elif isinstance(b, Pipit):
            b.membangun_sarang()
    else:
        print("Bukan burung")


if __name__ == "__main__":
    data = [
        Elang("Elang"),
        Pipit("Pipit"),
        Burung("Burung"),
        "kucing"
    ]

    for x in data:
        interaksi(x)
```

### Penugasan : Program dan Kelas diagram untuk hierarki kelas Komputer dan mendemonstrasikan  polimorfisme serta simulasi overloading

```python
# Kelas Induk
class Komputer:
    def __init__(self, merk, processor, ram_gb):
        self.merk = merk
        self.processor = processor
        self.ram_gb = ram_gb

    def info_spesifikasi(self):
        print(f"Merk      : {self.merk}")
        print(f"Processor : {self.processor}")
        print(f"RAM       : {self.ram_gb} GB")

    def jalankan_aplikasi(self, nama_aplikasi):
        print(f"{self.merk} menjalankan aplikasi: {nama_aplikasi}...")

    # Overloading (pakai *args)
    def install_software(self, *args):
        for software in args:
            print(f"Menginstall {software} di {self.merk}...")


# Kelas Anak Laptop
class Laptop(Komputer):
    def __init__(self, merk, processor, ram_gb, ukuran_layar_inch, berat_kg):
        super().__init__(merk, processor, ram_gb)
        self.ukuran_layar_inch = ukuran_layar_inch
        self.berat_kg = berat_kg

    def info_spesifikasi(self):
        super().info_spesifikasi()
        print(f"Ukuran Layar : {self.ukuran_layar_inch} inch")
        print(f"Berat        : {self.berat_kg} Kg")


# Kelas Anak Desktop
class Desktop(Komputer):
    def __init__(self, merk, processor, ram_gb, jenis_casing, monitor_external):
        super().__init__(merk, processor, ram_gb)
        self.jenis_casing = jenis_casing
        self.monitor_external = monitor_external

    def info_spesifikasi(self):
        super().info_spesifikasi()
        print(f"Jenis Casing     : {self.jenis_casing}")
        print(f"Monitor External : {'Ya' if self.monitor_external else 'Tidak'}")


# Fungsi Polimorfisme
def cetak_semua_spesifikasi(daftar_komputer):
    print("\n=== Daftar Spesifikasi Komputer ===")
    for komputer in daftar_komputer:
        print("----------------------------")
        komputer.info_spesifikasi()


# Kode Utama
if __name__ == "__main__":
    # Membuat objek
    laptop1 = Laptop("Asus", "Intel i7", 16, 15.6, 1.8)
    desktop1 = Desktop("HP", "AMD Ryzen 5", 8, "Tower", True)

    # List objek
    daftar = [laptop1, desktop1]

    # Polimorfisme
    cetak_semua_spesifikasi(daftar)

    print("\n=== Menjalankan Aplikasi ===")
    laptop1.jalankan_aplikasi("Visual Studio Code")

    print("\n=== Install Software (Overloading *args) ===")
    laptop1.install_software("Chrome")
    laptop1.install_software("Python", "Git", "NodeJS")
```
