```python
angka = int(input("Masukkan angka: "))

# Genap / Ganjil
if angka % 2 == 0:
    print("Genap")
else:
    print("Ganjil")

# Cek Prima
if angka > 1:
    for i in range(2, angka):
        if angka % i == 0:
            print("Bukan Prima")
            break
    else:
        print("Prima")
else:
    print("Bukan Prima")
```

```python
berat = float(input("Masukkan berat badan (kg): "))
tinggi = float(input("Masukkan tinggi badan (cm): ")) / 100

bmi = berat / (tinggi ** 2)

print("BMI =", bmi)

if bmi < 18.5:
    print("Underweight")
elif bmi < 25:
    print("Normal")
elif bmi < 30:
    print("Overweight")
else:
    print("Obesitas")
```
